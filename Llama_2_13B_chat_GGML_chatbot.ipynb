```python
model_name_or_path = "TheBloke/Llama-2-13B-chat-GGML"
model_basename = "llama-2-13b-chat.ggmlv3.q5_1.bin" # the model is in bin format
```

```python
from huggingface_hub import hf_hub_download
```

```python
from llama_cpp import Llama
```

```python
model_path = hf_hub_download(repo_id=model_name_or_path, filename=model_basename)
```

```python
# GPU
lcpp_llm = None
lcpp_llm = Llama(
    model_path=model_path,
    n_threads=2, # CPU cores
    n_batch=512, # Should be between 1 and n_ctx, consider the amount of VRAM in your GPU.
    n_gpu_layers=0 # Change this value based on your model and your GPU VRAM pool.
    )
```

```python
prompt = "write a python function to sum 2 numbers"
prompt_template=f'''SYSTEM: You are a helpful python coding asssitant.

USER: {prompt}

ASSISTANT:
'''
```

```python
response=lcpp_llm(prompt=prompt_template, max_tokens=256, temperature=0.5, top_p=0.95,
                  repeat_penalty=1.2, top_k=150,
                  echo=True)
```

```python
response = response["choices"][0]["text"].split("ASSISTANT:")[1]
```

```python
print(response)
```

```python
def create_chatbot_response(prompt):
  prompt_template=f'''SYSTEM: You are a helpful python coding asssitant.

  USER: {prompt}

  ASSISTANT:
  '''
  response = lcpp_llm(prompt=prompt_template, max_tokens=256, temperature=0.5, top_p=0.95,
                      repeat_penalty=1.2, top_k=150,
                      echo=True)
  response = response["choices"][0]["text"]
  return response

while True:
  # Get user input
  prompt = input("USER: ")

  # Check for quit command
  if prompt.lower() == "quit":
    break

  # Generate response if not quit
  response = create_chatbot_response(prompt)
  print(response)

print("Chatbot stopped.")
```
